```python
# 강사님 코드
def calculate_max_credits(lectures):
    # 강의를 끝나는 시간 기준으로 정렬
    sorted_lectures = sorted(lectures, key=lambda x: x["end"])
    max_credits = 0
    last_end_time = "00:00"

    for lecture in sorted_lectures:
        # 이전 강의가 끝난 시간과 현재 강의의 시작 시간을 비교
        if lecture["start"] >= last_end_time:
            max_credits += lecture["credit"]
            last_end_time = lecture["end"]

    return max_credits
```

```python
def solution(data):
    # 주어진 항공편 정보를 변수에 저장합니다.
    flights = data

    # Flight 클래스를 정의합니다.
    class Flight:
        def __init__(self, name, start, end):
            # 항공편의 이름, 출발 시간, 도착 시간을 인스턴스 변수로 설정합니다.
            self.name = name
            self.start = start
            self.end = end

        # '<' 연산자 오버로딩을 통해 Flight 인스턴스들을 도착 시간을 기준으로 비교할 수 있도록 합니다.
        def __lt__(self, other):
            return self.end < other.end

    # 문자열 정보를 기반으로 Flight 인스턴스 리스트를 생성합니다.
    flight_objects = []
    for flight in flights:
        # 각 항공편 정보를 ','를 기준으로 분리하여 이름, 출발 시간, 도착 시간을 얻습니다.
        name, start, end = flight.split(",")
        # Flight 클래스의 인스턴스를 생성하고 리스트에 추가합니다.
        flight_objects.append(Flight(name, start, end))

    # 도착 시간을 기준으로 항공편을 정렬합니다.
    flight_objects.sort()

    # 가장 빠른 항공편의 이름을 반환합니다.
    return flight_objects[0].name
```

```python
# 85
# 추천 수가 가장 많은 코드
def solution(data):
    airports, patientData = data

    n_dict = {i["name"]:(i["distance"], i["treatableDiseases"]) for i in airports}
    air_list = list()

    n_d = sorted(n_dict.items() ,key = lambda x: x[1][0])

    for i in n_d:
        if i[1][1] == patientData:
            return [i[0]]

    for i in n_d:
        for treat in i[1][1]:
            if treat in patientData:
                patientData.remove(treat)
                air_list.append(i[0])

    return air_list
```

```python
# 86번
def solution(data):
    # 문자열에서 각 문자의 개수 계산
    char_counts = {}
    for char in data:
        if char in char_counts:
            char_counts[char] += 1
        else:
            char_counts[char] = 1

    # 개수가 많은 순서대로 정렬, 같으면 알파벳 순서대로 정렬
    sorted_chars = sorted(char_counts.items(), key=lambda x: (-x[1], x[0]))

    # 중복 제거된 문자열 생성
    result = ''.join(char for char, count in sorted_chars)

    return result
```

```python
def solution(s):
    frequency = {}
    for char in s:
        if char.isalpha():  # 알파벳인 경우에만 처리
            frequency[char] = frequency.get(char, 0) + 1


    sorted_alphabets = sorted(frequency.items(), key=lambda x: (-x[1], x[0]))


    result = "".join([char for char, _ in sorted_alphabets])
    return result
```

```python
# 추천 수가 가장 많은 코드 수
def solution(data):
    my_dict = {0: '', 1: '일', 2: '이', 3:'삼', 4:'사', 5:'오',
               6:'육', 7:'칠', 8:'팔', 9:'구', 10:'십', 100:'백', 1000:'천'}
    result = ''
    current_num = data
    if current_num // 1000 != 0:
        result += my_dict[data // 1000] + '천'
        current_num = data % 1000
    if current_num // 100 != 0:
        result += my_dict[current_num // 100] + '백'
        current_num = data % 100
    if current_num // 10 != 0:
        result += my_dict[current_num // 10] + '십'
        current_num = data % 10
    result += my_dict[current_num]
    return result
```

```python
# 88번
def solution(data):
    numbers = ["", "일", "이", "삼", "사", "오", "육", "칠", "팔", "구"]
    units = ["", "십", "백", "천"]
    result = []
    str_data = str(data)

    for i, num in enumerate(str_data):
        num = int(num)
        unit = units[len(str_data) - i - 1]

        if num != 0 or unit == "":
            result.append(numbers[num] + unit)

    # 첫 자리가 1인 경우 특별 처리
    if len(result) > 1 and result[0] == "일":
        result = result[1:]

    return "".join(result)
```
