```python
# 48번
def solution(data):
    tree = data["tree"]

    # 모든 경로의 합을 저장할 리스트
    paths_sum = []

    def bin_tree(node, current_sum=0):
        if not node:
            return

        # 현재 노드의 value를 더해줌
        current_sum += node["value"]

        # 리프 노드에 도달한 경우, 현재까지의 합을 paths_sum에 추가
        if "left" not in node and "right" not in node:
            paths_sum.append(current_sum)
            return

        # left 자식 노드가 있는 경우, 재귀적으로 탐색
        if "left" in node:
            bin_tree(node["left"], current_sum)

        # right 자식 노드가 있는 경우, 재귀적으로 탐색
        if "right" in node:
            bin_tree(node["right"], current_sum)

    # 이진 트리 탐색 시작
    bin_tree(tree)

    return paths_sum
```

# 문제 49번
### 그래프의 최단 경로 찾기

```python
# 풀어낸 사람이 없었음. 강사님 코드
# 49번
from collections import deque


def solution(data):
    def bfs_shortest_path(graph, start, end):
        visited = set()
        queue = deque([(start, 0)])  # (current node, distance)

        while queue:
            current, distance = queue.popleft()
            if current == end:
                return distance

            if current not in visited:
                visited.add(current)
                for neighbor in graph.get(current, []):
                    queue.append((neighbor, distance + 1))

        return -1  # Path not found

    graph = data["graph"]
    start = data["start"]
    end = data["end"]
    return bfs_shortest_path(graph, start, end)
```

# 문제 50번
### 그래프 사이클 감지

```python
# 풀어낸 사람이 없었음. 강사님 코드
# 50번
def solution(data):
    def has_cycle(graph):
        visited = set()
        rec_stack = set()

        for node in graph:
            if node not in visited:
                if dfs(graph, node, visited, rec_stack):
                    return True

        return False

    def dfs(graph, current, visited, rec_stack):
        if current not in visited:
            visited.add(current)
            rec_stack.add(current)

            for neighbor in graph.get(current, []):
                if neighbor not in visited:
                    if dfs(graph, neighbor, visited, rec_stack):
                        return True
                elif neighbor in rec_stack:
                    return True

        rec_stack.remove(current)
        return False

    return has_cycle(data["graph"])
```

# 문제 51번
### 최소 동전 개수 찾기

```python
# 가장 추천을 많이 받은 코드 - 6
# 51번
def solution(data):
    coins = sorted(data['coins'], reverse=True)
    amount = data['amount']

    num_coin = 0
    for coin in coins:
        while amount >= coin:
            amount -= coin
            num_coin += 1
    return num_coin
```

```python
# 51번
def solution(data):
    coins, amount, count = data["coins"], data["amount"], 0
    coins = sorted(coins, reverse=True)
    for coin in coins:
        if coin <= amount:
            num, amount = divmod(amount, coin)  # 몫과 나머지
            count += num
        if amount == 0:
            return count
    return -1  # 주어진 동전들로 amount를 만들 수 없는 경우
```
